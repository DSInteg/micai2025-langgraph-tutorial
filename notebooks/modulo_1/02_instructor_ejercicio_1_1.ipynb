```python
# IMPLEMENTACIÃ“N PASO A PASO

from typing import TypedDict
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI

# Paso 1: Estado
class DocumentState(TypedDict):
    document: str
    cleaned_text: str
    sentiment: str
    summary: str

print("âœ… Estado definido")
```

```python
# Paso 2: Nodos

llm = ChatOpenAI(model='gpt-4o-mini')

def clean_text_node(state):
    text = state['document'].strip().lower()
    print(f"ðŸ§¹ Limpiando: {text[:50]}...")
    return {'cleaned_text': text}

def analyze_sentiment_node(state):
    response = llm.invoke(f"Sentiment (Positivo/Negativo/Neutral): {state['cleaned_text'][:200]}")
    sentiment = response.content.strip()
    print(f"ðŸŽ­ Sentimiento: {sentiment}")
    return {'sentiment': sentiment}

def summarize_node(state):
    response = llm.invoke(f"Resume en una oraciÃ³n: {state['cleaned_text'][:200]}")
    summary = response.content.strip()
    print(f"ðŸ“ Resumen: {summary}")
    return {'summary': summary}

print("âœ… Nodos definidos")
```

```python
# Paso 3: Construir Grafo

workflow = StateGraph(DocumentState)

workflow.add_node('clean', clean_text_node)
workflow.add_node('analyze', analyze_sentiment_node)
workflow.add_node('summarize', summarize_node)

workflow.add_edge(START, 'clean')
workflow.add_edge('clean', 'analyze')
workflow.add_edge('analyze', 'summarize')
workflow.add_edge('summarize', END)

app = workflow.compile()

print("âœ… Workflow compilado")
```

```python
# ðŸŽ¬ DEMO: Ejecutar

test_doc = 'LangGraph es increÃ­ble para construir workflows!'

print('='*60)
print('ðŸŽ¬ EJECUTANDO WORKFLOW')
print('='*60)
print(f'Input: {test_doc}\n')

result = app.invoke({'document': test_doc})

print('\nðŸ“Š RESULTADO:')
for key, value in result.items():
    print(f'  {key}: {value}')
print('='*60)
```
